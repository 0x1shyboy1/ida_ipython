# Snaphots with Sark
With the help of [Sark](sark.rtfd.org), IDA-IPython can become a great tool for documenting important parts of your work and your IDB, as Sark allows you to easily take snapshots of IDA.

```python
import sark
from sark.ipython import snap
```

```python
snap('IDA View-A')
```

It also possible to capture the entire IDA screen.

```python
snap()
```
